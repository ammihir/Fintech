# Toronto Dwellings Analysis Dashboard

In this notebook, you will compile the visualizations from the previous analysis into functions to create a Panel dashboard.

```python
# initial imports
import os
import pandas as pd
import matplotlib.pyplot as plt
import hvplot.pandas
import plotly.express as px
from pathlib import Path
from dotenv import load_dotenv

# hides matplotlib graphs
# %matplotlib agg
```

```python
# Initialize the Panel Extensions (for Plotly)
import panel as pn
pn.extension("plotly")
```

```python
# Read the Mapbox API key
load_dotenv()
map_box_api = os.getenv("mapbox")
px.set_mapbox_access_token(map_box_api)
```

# Import Data

```python
# Import the CSVs to Pandas DataFrames
file_path = Path("Data/toronto_neighbourhoods_census_data.csv")
to_data = pd.read_csv(file_path, index_col="year")

file_path = Path("Data/toronto_neighbourhoods_coordinates.csv")
df_neighbourhood_locations = pd.read_csv(file_path, index_col='neighbourhood')
```

## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

### Global available data

```python
# Getting the data from the top 10 expensive neighbourhoods
top_10_expensive = (to_data.groupby('neighbourhood')
                    .mean()
                    .sort_values('average_house_value',ascending=False)
                    .reset_index()
                    .head(10)
                   )

# Calculate the mean number of dwelling types units per year
dwellings = ['single_detached_house', 'apartment_five_storeys_plus',
       'movable_dwelling', 'semi_detached_house', 'row_house', 'duplex',
       'apartment_five_storeys_less', 'other_house']

to_by_dwell_type = (to_data
                  .loc[:,dwellings]
                  .groupby(to_data.index)
                  .sum()
                  )

# Calculate the average monthly shelter costs for owned and rented dwellings
avg_monthly_shelter = (to_data.loc[:,['shelter_costs_owned', 'shelter_costs_rented']]
                       .groupby(to_data.index)
                       .mean()
                      ).reset_index()
```

```python
# Define Panel visualization functions
def neighbourhood_map():
    """Neighbourhood Map"""
    
    avg_neigh_list = ['single_detached_house', 'apartment_five_storeys_plus',
       'movable_dwelling', 'semi_detached_house', 'row_house', 'duplex',
       'apartment_five_storeys_less', 'other_house', 'average_house_value',
       'shelter_costs_owned', 'shelter_costs_rented']
    
    avg_neighbourhood = (to_data.groupby('neighbourhood')
                    .mean()[avg_neigh_list]
                        )
    
    avg_neighb_loc_df = pd.concat([df_neighbourhood_locations,avg_neighbourhood],axis='columns').reset_index()
    
    house_map = px.scatter_mapbox(
        avg_neighb_loc_df,
        lat="lat",
        lon="lon",
        color="average_house_value",
        size='average_house_value',
        title='Average House Values in Toronto',
        zoom=9.5,
        hover_name='neighbourhood',
        hover_data=avg_neighb_loc_df.columns,
        height=600,
        color_continuous_scale=px.colors.cyclical.IceFire
    )
    
    return house_map
     

def create_bar_chart(data, title, xlabel, ylabel, color):
    """
    Create a barplot based in the data argument.
    """    
    
    fig = plt.figure(figsize=(6,6))
    
    plt.bar(data=data,
           x=data.index,
           height=data,
           color=color)

    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.xticks(rotation=90)
    
    plt.tight_layout()
    plt.close(fig)
    
    return fig

def create_line_chart(data, title, xlabel, ylabel, color):
    """
    Create a line chart based in the data argument.
    """
    fig = plt.figure()
    
    plt.plot(data,color=color)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    
    plt.tight_layout()
    plt.close(fig)
    
    return fig

def average_house_value():
    """Average house values per year."""
    
    average_house_value = (to_data.loc[:,'average_house_value']
                      .groupby(to_data.index)
                      ).mean()
    
    return create_line_chart(average_house_value, 
                  'Average House Value per Year',
                  'Year',
                  'Average House Value',
                  'blue') 

def average_value_by_neighbourhood():
    """Average house values by neighbourhood."""
    
    mean_house_value_by_neigh = (to_data.loc[:,['neighbourhood','average_house_value']]
                            .reset_index())
    
    return mean_house_value_by_neigh.hvplot(x='year',
                                 groupby='neighbourhood',
                                 width=750,
                                 height=500)

def number_dwelling_types():
    """Number of dwelling types per year"""
    
    neighb_dwell_list = ['single_detached_house', 'apartment_five_storeys_plus',
       'movable_dwelling', 'semi_detached_house', 'row_house', 'duplex',
       'apartment_five_storeys_less', 'other_house']
    
    dwell_per_neighb = (to_data
        .groupby(['neighbourhood','year'])[neighb_dwell_list]
        .sum()
       ).reset_index()
    
    return dwell_per_neighb.hvplot.bar(x='year',
                            ylabel='Dwelling Type Units',
                            xlabel='Year',
                            groupby='neighbourhood',
                            width=750,
                            height=500,
                            rot=90)
    
def average_house_value_snapshot():
    """Average house value for all Toronto's neighbourhoods per year."""
    
    cost_analysis = pd.DataFrame(to_data
                .groupby(['year','neighbourhood'])
                .sum()['average_house_value']
                .reset_index()
                )
    
    
    
    fig = px.bar(cost_analysis, x="neighbourhood", 
             y='average_house_value',
             color='average_house_value', 
             facet_row="year",
             height=1000,
             width=1500,
             title='Average House Values in Toronto per Neighbourhood'
            ) 
    fig.update_yaxes(matches=None)
    
    return fig

def top_most_expensive_neighbourhoods():
    """Top 10 most expensive neighbourhoods."""
    return top_10_expensive.hvplot.bar(x='neighbourhood',
                                y='average_house_value',
                                rot=90, 
                                height=500,
                                width=750, 
                                title='Top 10 Expensive Neighbourhoods in Toronto',
                                ylabel='Average House Value',
                                xlabel='Neighbourhood')

def sunburts_cost_analysis():
    """Sunburst chart to conduct a costs analysis of most expensive neighbourhoods in Toronto per year."""
    
    expensive_neighbours = set(top_10_expensive['neighbourhood'])
    
    sunburst_cost_analysis = (to_data
                              .loc[to_data['neighbourhood'].isin(expensive_neighbours)]
                             ).reset_index()
    
    sun_fig = px.sunburst(data_frame=sunburst_cost_analysis,
                          path=['year','neighbourhood'],
                          values='shelter_costs_owned',
                          color='shelter_costs_owned',
                          color_continuous_scale="blues",
                          title='Cost Analysis of Most Expensive Neighbourhoods in Toronto per Year',
                          hover_data=['shelter_costs_rented','average_house_value']
                         )
    return sun_fig
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# Create a Title for the Dashboard
Welcome = '# Real Estate Analysis of Toronto from 2001 to 2016'


# Layouts in each tab
## Introduction Tab
introduction_text = pn.Column(
    "The following real estate data shows the progression of Toronto's housing market from 2001 to 2016.",
                         neighbourhood_map()
                        )

## Yearly Market Analysis Tab
yearly_analysis = pn.Column(
    pn.Row(create_bar_chart(to_by_dwell_type.loc[2001], 
                               "Dwelling Types in Toronto in 2001", 
                                "2001", 
                                "Dwelling Type Units", 
                                "red"
                              ),
              create_bar_chart(to_by_dwell_type.loc[2006], 
                               "Dwelling Types in Toronto in 2006", 
                                "2006", 
                                "Dwelling Type Units", 
                                "blue"
                              )
             ),
    pn.Row(create_bar_chart(to_by_dwell_type.loc[2011], 
                               "Dwelling Types in Toronto in 2011", 
                               "2011", 
                               "Dwelling Type Units", 
                               "orange"
                              ),
              create_bar_chart(to_by_dwell_type.loc[2016], 
                               "Dwelling Types in Toronto in 2016", 
                               "2016", 
                               "Dwelling Type Units",
                               "magenta"
                              )
             )
)


## Shelter Costs vs. House Value Tab
shelter_panel = pn.Row(create_line_chart(avg_monthly_shelter["shelter_costs_rented"],
                         "Average Monthly Shelter Cost for Rented Dwellings in Toronto",
                         "Year",
                         "Avg Monthly Shelter Costs", 
                         "orange"
                        ),
       create_line_chart(avg_monthly_shelter["shelter_costs_owned"],
                         "Average Monthly Shelter Cost for Owned Dwellings in Toronto",
                         "Year",
                         "Avg Monthly Shelter Costs", 
                         "blue"
                        )
      )

## Neighbourhood Analysis Tab
neighbourhood_analysis = pn.Column(number_dwelling_types(),
                                   average_value_by_neighbourhood(),
                                   average_house_value_snapshot()
                                  )

## Top Expensive Neighbourhoods Tab
top_expensive_neighb = pn.Column(sunburts_cost_analysis(),
                              top_most_expensive_neighbourhoods()
                             )

Loui = pn.Column("This is Loui. I am dog sitting him. Please let me know in your grading comments which two legs are his front legs. I am very concerned.",pn.panel('Data/Loui.png', width=300))

# Create a tab layout for the dashboard
dashboard = pn.Column(Welcome,
                      pn.Tabs(
                          ("Introduction", introduction_text),
                          ("Yearly Market Analysis",yearly_analysis),
                          ('Shelter Costs vs. House Value',shelter_panel),
                          ('Neighbourhood Analysis',neighbourhood_analysis),
                          ('Top Expensive Neighbourhoods',top_expensive_neighb),
                          ('Loui',Loui)
                      )
                     )
```

## Serve the Panel Dashboard

```python
dashboard.servable().show()
```
